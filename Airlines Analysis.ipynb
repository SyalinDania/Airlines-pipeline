```python
# Import the pandas library.
import pandas
import seaborn as sns
import numpy as np
# Read in the airports data.
airports = pandas.read_csv("airports.txt", header=None, dtype=str)
```

```python
airports.shape
```

```python
airports.head()
```

### Data Pre-processing

1) Remove last 2 columns and insert column names

```python
airports=airports.iloc[:,0:12]
```

```python
airports.columns = ["ID", "Name", "City", "Country", "IATA", 
                    "ICAO", "Latitude", "Longitude", 
                    "Altitude", "Offset", "DST", "Timezone"]
airports.head()
```

```python
airports.head()
```

```python
airports.isnull()
```

```python
airports.isnull().values.any()
```

```python
airports.isnull().sum()
```

```python
airports_new = airports.dropna()
```

```python
airports_new.head()
```

```python
airports_new.isnull().sum()
```

```python
sns.heatmap(airports_new.isnull(),yticklabels=False)
```

```python
airports_new.to_csv('airports_updated.csv', index=False)
```

```python
airports_new.shape
```

```python
airports_new.tail()
```

```python
airports_new.info()
```

2) Change Data Types 

```python
airports_new[['ID', 'Altitude']] = airports_new[['ID', 'Altitude']].astype(int)
```

```python
airports_new[['Latitude','Longitude']] = airports_new[['Latitude','Longitude']].astype(float)
```

```python
airports_new.info()
```

```python
airports_new.head()
```

```python
airports_new.dropna(axis='index', how='all')
```

## Cont. Data Pre-processing : Drop \N values

```python
airports_new.replace("\\N",np.nan,inplace=True)
```

```python
airports_final =airports_new.dropna()
```

```python
airports_final.tail()
```

```python
airports_final.shape
```

```python
sns.heatmap(airports_final.isnull(),yticklabels=False)
```

```python
airports_final[['Offset']] = airports_new[['Offset']].astype(float)
```
